```python
import pandas as pd 
import pandas as pd

df1 = pd.DataFrame({
    'Emp_ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Department': ['HR', 'IT', 'Finance']
})
df2 = pd.DataFrame({
    'Employee_Number': [2, 3, 4],
    'Salary': [50000, 60000, 70000]
})
merge = pd.merge(df1,df2,how='outer',left_on='Emp_ID',right_on='Employee_Number')
merge
```

```python
#handling missing
mean = merge['Salary'].mean()
merge.fillna({
    'Emp_ID':4.0,
    'Name':'Unknown',
    'Department':'Unknown',
    'Salary':mean
},inplace=True)
merge
```
