```python
import pandas as pd 
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]}, index=[0, 1])
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]}, index=[2, 3])
concat = pd.concat([df1,df2], ignore_index=True)
concat
```

```python
#outer join adds all the values in Both dataFrames and out NaN where the values are not common
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'B': [5, 6], 'C': [7, 8]})
concat2 = pd.concat([df1,df2], ignore_index=True, join='outer')
concat2
```

```python
#inner join only takes the values which are common in both dataFrames
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'B': [5, 6], 'C': [7, 8]})
concat2 = pd.concat([df1,df2], ignore_index=True, join='inner')
concat2
```

```python
#handling missing values after using outer Joinabs
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]}, index=[0, 1])
df2 = pd.DataFrame({'B': [5, 6], 'C': [7, 8]}, index=[2, 3])
con = pd.concat([df1,df2],ignore_index=True,join='outer')
con
```

```python
#using key arg and putting key horizontally (for vertical use axis=0 or leave it as is because by default it's veritcal)
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
con1 = pd.concat([df1,df2],join='outer',axis=1,keys=['df1','df2'])
con1
```
