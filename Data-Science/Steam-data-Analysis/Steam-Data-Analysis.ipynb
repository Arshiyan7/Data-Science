# Handling Missing Values  

In this section, we'll identify and handle missing values using appropriate strategies like filling or dropping based on the nature of the data.


```python
def fill_missing_strings(df, fill_dict):
    for column, fill_value in fill_dict.items():
        if column in df.columns:
            df[column] = df[column].fillna(fill_value)
    return df


fill_values = {
    'types': 'unknown',
    'name': 'not-provided',
    'desc_snippet': 'no_description',
    'recent_reviews': 'no_reviews',
    'all_reviews': 'No recent reviews (0)',
    'developer': 'unknown developer',
    'publisher': 'unknown publisher',
    'popular_tags': 'no tags provided',
    'game_details': 'no details provided',
    'languages': 'English',
    'genre': 'Action',
    'game_description': 'no Description provided',
    'mature_content': 'no details',
    'minimum_requirements': 'no_minimum requirements',
    'recommended_requirements': 'not listed',
    'original_price': 0
}

data = fill_missing_strings(copy_data, fill_values)
```

### Note:
To handle missing values effectively, we assigned default values to specific columns based on their most frequently occurring entries:

- **Language**: Filled with `"English"` as it was the most commonly listed language.
- **Genre**: Filled with `"Action"` since it appeared most frequently in dataset
s.


### Handling Missing Release Dates

To address the missing values in the **`release_date`** column, we replaced them **randomly with dates from the years 2016 and 2017**. These years were selected based on their high frequency in the dataset. 

Using this method helps preserve the overall distribution of the data without skewing it toward a single placeholder year like `1900` or `2000`, and maintains a more natural variation in the release timeline.


```python
import random
replacement_years = [2016, 2017]
missing_mask = copy_data['release_date'].isna()

random_dates = pd.to_datetime(
    [f"{random.choice(replacement_years)}-01-01" for _ in range(missing_mask.sum())]
)
copy_data.loc[missing_mask, 'release_date'] = random_dates

print(f"Missing values remaining in release_date: {copy_data['release_date'].isnull().sum()}")
```

```python
print("\n--- Verifying Missing Value Removal ---")
remaining_missing_total = copy_data.isna().sum().sum()

if remaining_missing_total == 0:
    print("Success! All missing values have been handled across the dataset.")
else:
    print(f"Warning! There are still {remaining_missing_total} missing values remaining.")
    print("Columns with remaining missing values:")

    print(copy_data.isna().sum()[copy_data.isna().sum() > 0])

print("-------------------------------------\n")
print(f"\nShape of the cleaned dataset: {copy_data.shape}")
```

```python
def handle_repeat_publisher(df, column):
    for col in column:
        df[col] = df[col].astype(str)
        df[col] = df[col].str.split(',').apply(lambda x: [i.strip() for i in x])
        df[col] = df[col].apply(lambda x: ', '.join(sorted(set(x), key=x.index)))
    return df

val = ['publisher']
copy_data = handle_repeat_publisher(copy_data, val)
```

```python
import re

def clean_price_column(column):
    cleaned_prices = []
    for val in column:
        if pd.isna(val):
            cleaned_prices.append(0)
            continue
        val = str(val).lower()
        if 'free' in val:
            cleaned_prices.append(0)
        elif re.search(r'\d', val):  # Check if it contains at least one digit
            val = val.replace('$', '').strip()
            try:
                cleaned_prices.append(float(val))
            except ValueError:
                cleaned_prices.append(0)
        else:
            cleaned_prices.append(0)
    return cleaned_prices

copy_data['original_price'] = clean_price_column(copy_data['original_price'])
copy_data['discount_price'] = copy_data['discount_price'].fillna(copy_data['original_price'])
```

###  Price Columns Cleaning

- Replaced 'Free' and 'Free to Play' in `original_price` with 0.
- Removed '$' symbols and unwanted text from both `original_price` and `discount_price`.
- Converted both columns to `float` type.
- Filled missing `discount_price` values using corresponding `original_price`.
