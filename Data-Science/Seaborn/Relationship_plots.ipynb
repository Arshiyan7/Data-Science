```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
#Load the dataset 
crash_df = sns.load_dataset('car_crashes')

# Display dataset info
print("="*40)
print("üìå DATASET INFORMATION")
print("="*40)
print(crash_df.info())

# Display first 5 rows
print("\n" + "="*40)
print("üîç FIRST 5 ROWS OF THE DATASET")
print("="*40)
print(crash_df.head())

# Display last 5 rows
print("\n" + "="*40)
print("üìå LAST 5 ROWS OF THE DATASET")
print("="*40)
print(crash_df.tail())

# Display basic statistics
print("\n" + "="*40)
print("üìä BASIC STATISTICS")
print("="*40)
print(crash_df.describe())
```

```python
#Join plot 
plt.figure(figsize=(5,2))
sns.set_style('whitegrid')
sns.jointplot(data=crash_df,x='speeding',y='total',kind='reg',color='navy')
```

```python
# Pair Plot  
sns.set_style("whitegrid")  
sns.pairplot(crash_df, dropna=True, palette="coolwarm", hue="abbrev") 
```

```python
plt.figure(figsize=(8, 5))
sns.set_style("whitegrid")
sns.rugplot(data=crash_df, x="speeding", color="darkred", height=0.1, linewidth=1, alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

sns.swarmplot(data=crash_df, x="alcohol", y="total", palette="coolwarm", size=6,hue=crash_df['alcohol'])

# Using Matplotlib for customization
plt.title("Swarm Plot of Alcohol Influence on Total Accidents", fontsize=14, fontweight='bold')
plt.xlabel("Alcohol Consumption Level", fontsize=12)
plt.ylabel("Total Accidents", fontsize=12)

plt.show()
```
