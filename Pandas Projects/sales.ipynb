```python
import pandas as pd 
df = pd.read_csv('sales_data.csv')
df
```

```python
#inspecting dataset
copy = df.copy()
print ("First five rows of dataset:\n ",copy.head(5))
print("\nMissing values total in each column:\n", copy.isna().sum())
print("\nData types of each columns: ",copy.dtypes)
```

```python
copy['Date'] = pd.to_datetime(copy['Date'])
print(copy['Date'].dtype)
missing = copy.dropna(subset=['Date','CustomerID']).fillna({
    'Quantity':copy['Quantity'].mean(),
    'TotalAmount':copy['TotalAmount'].median(),
    'UnitPrice':copy['UnitPrice'].max()
})
missing.isna().sum()
```

```python
#verification of amounts
missing['Quantity']*missing['UnitPrice'] 
```

```python
#plotting sales by week:
import matplotlib.pyplot as plt 
import matplotlib.dates as mdates
weekly_sales = copy.resample('W', on='Date')['TotalAmount'].sum()
ax = weekly_sales.plot(kind='bar',figsize=(10,6),title="total sales over Time (Daily)",xlabel="Dates",ylabel="Total sales",color='navy')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d/%m/%Y')) 
plt.xticks(rotation=0)
plt.show()
```

```python
#plotting sales by day 
#plotting sales by week:
import matplotlib.pyplot as plt 
import matplotlib.dates as mdates
daily_sales = copy.resample('D', on='Date')['TotalAmount'].sum()
ax = daily_sales.plot(kind='bar',figsize=(10,6),title="total sales over Time (Daily)",xlabel="Dates",ylabel="Total sales",color='red')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d/%m/%Y')) 
plt.xticks(rotation=60)
plt.show()
```

```python
#plotting sales on month:
#plotting sales by week:
import matplotlib.pyplot as plt 
import matplotlib.dates as mdates
monthly_sales = copy.resample('ME', on='Date')['TotalAmount'].sum()
ax = monthly_sales.plot(kind='bar',figsize=(10,6),title="total sales over Time (Daily)",xlabel="Dates",ylabel="Total sales",color='chartreuse')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d/%m/%Y')) 
plt.xticks(rotation=0)
plt.show()
```

```python
#Sales by product:
sales_byP = copy.groupby('Product')['Quantity'].sum()
sales_byP.plot(kind='bar',figsize=(10,6),title="sales by product",xlabel='Product',ylabel='Product',color='b')
plt.xticks(rotation=0)
```

```python
copy['Quantity'].plot(kind='hist', bins=20, alpha=0.7, color='blue', edgecolor='black',figsize=(10,6))
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.show()
```
