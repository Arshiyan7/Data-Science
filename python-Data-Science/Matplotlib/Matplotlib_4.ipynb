```python
import matplotlib.pyplot as plt
import numpy as np

companies = ["Apple", "Microsoft", "Amazon", "Google", "Tesla", ]
revenues = [394.3, 211.9, 469.8, 283.2, 81.5, ]
```

```python
xpos = np.arange(len(companies))
```

```python
plt.bar(xpos,revenues,label='revenues',color='r')
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
profit = [32,31,45,32,55]
plt.bar(xpos-0.2,revenues,width=0.4,label='revenues',color='r')
plt.bar(xpos+0.2,profit,width=0.4,label='profit',color='g')
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```
