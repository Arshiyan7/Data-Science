```python
#Create a copy of dataset so the change reflect 
df_cleaned = df.copy()
```

```python
def handling_missing_data(row):
    if pd.isna(row['name']):
        row['name'] = 'Unknown'

    if pd.isna(row['types']):
        row['types'] = 'not_mentioned'
     
    if pd.isna(row['desc_snippet']):
        row['desc_snippet'] = 'no_description'

    if pd.isna(row['all_reviews']):
        row['all_reviews'] = 'no_reviews'

    if pd.isna(row['recent_reviews']):
        row['recent_reviews'] = 'no_reviews'

    if pd.isna(row['developer']):
        row['developer'] = 'No_Info'

    if pd.isna(row['publisher']):
        row['publisher'] = 'No_Info'

    if pd.isna(row['popular_tags']):
        row['popular_tags'] = 'Unknown'

    if pd.isna(row['game_details']):
        row['game_details'] = 'Not_Mentioned'

    if pd.isna(row['languages']):
        row['languages'] = 'English' #English is by default for all games 

    if pd.isna(row['genre']):
        row['genre'] = 'Unknown' #Action Genre is common

    if pd.isna(row['game_description']):
        row['game_description'] = 'Not_Provided'

    if pd.isna(row['mature_content']):
        row['mature_content'] = 'Not_Provided'

    if pd.isna(row['minimum_requirements']):
        row['minimum_requirements'] = 'Not_Specified'

    if pd.isna(row['recommended_requirements']):
        row['recommended_requirements'] = 'Not_Specified'

    return row

df_cleaned = df.apply(handling_missing_data, axis=1)

df_cleaned.isna().sum()
```

```python
df_cleaned.replace('', np.nan, inplace=True)
df_cleaned.drop(['achievements', 'original_price', 'discount_price'], axis=1, inplace=True)
```

```python
df_cleaned.isna().sum()
```

```python
df_cleaned['release_date'] = pd.to_datetime(df_cleaned['release_date'], format='mixed', dayfirst=True, errors='coerce')
df_cleaned['release_date'] = df_cleaned['release_date'].fillna(pd.to_datetime('1900-01-01'))
df_cleaned['release_date'].isna().sum()

df_cleaned['yearly_release'] = df_cleaned['release_date'].dt.year
df_cleaned['monthly_release'] = df_cleaned['release_date'].dt.month
df_cleaned['yearly_release'] = df_cleaned['yearly_release'].astype(int)
df_cleaned['monthly_release'] = df_cleaned['monthly_release'].astype(int)
```

```python
df_cleaned['popular_tags'].value_counts()
```

```python
def clean_text_column(df, column_list):
    for column in column_list:
        df[column] = df[column].str.lower().str.strip()
    return df

text_columns = ['genre', 'developer', 'publisher', 'popular_tags', 'game_details', 'languages', 'game_description', 'mature_content', 'minimum_requirements', 'recommended_requirements']
df_cleaned = clean_text_column(df_cleaned, text_columns)

```

```python
def add_count_columns(df, columns):
    for col in columns:
        df[f'{col}_count'] = df[col].apply(len)
    return df

list_cols = ['genre', 'popular_tags', 'game_details']
df_cleaned = add_count_columns(df_cleaned, list_cols)

df_cleaned
```

```python
def process_multi_label_columns(df, columns):
    for col in columns:
        list_col = f"{col}_list"
        count_col = f"{col}_count"

        df[list_col] = df[col].fillna("").apply(lambda x: [item.strip() for item in x.split(',')] if isinstance(x, str) else [])
        df[count_col] = df[list_col].apply(len)

    return df

multi_label_cols = ['genre', 'popular_tags', 'game_details']
df_cleaned = process_multi_label_columns(df_cleaned, multi_label_cols)
```
