```python
#Create a copy of dataset so the change reflect 
df_cleaned = df.copy()
```

```python
df_cleaned.dtypes #before date was object
df_cleaned['release_date'] = pd.to_datetime(df_cleaned['release_date'], format='mixed', dayfirst=True, errors='coerce')
df_cleaned.dtypes #After changing date to datetime
```

```python
#Drop the null columns in type column 
df_cleaned.dropna(subset=['types'],inplace=True)
df_cleaned['types'].isna().sum()
```

```python
def handling_missing_data(row):
    if pd.isna(row['name']):
        row['name'] = 'Unknown'
     
    if pd.isna(row['desc_snippet']):
        row['desc_snippet'] = 'no_description'

    if pd.isna(row['all_reviews']):
        row['all_reviews'] = 'no_reviews'

    if pd.isna(row['recent_reviews']):
        row['recent_reviews'] = 'no_reviews'

    if pd.isna(row['release_date']):
        row['release_date'] = 'Unknown release date'

    if pd.isna(row['developer']):
        row['developer'] = 'No_Info'

    if pd.isna(row['publisher']):
        row['publisher'] = 'No_Info'

    if pd.isna(row['popular_tags']):
        row['popular_tags'] = 'Unknown'

    if pd.isna(row['game_details']):
        row['game_details'] = 'Not_Mentioned'

    if pd.isna(row['languages']):
        row['languages'] = 'English' #English is by default for all games 

    if pd.isna(row['genre']):
        row['genre'] = 'Unknown' #Action Genre is common

    if pd.isna(row['game_description']):
        row['game_description'] = 'Not_Provided'

    if pd.isna(row['mature_content']):
        row['mature_content'] = 'Not_Provided'

    if pd.isna(row['minimum_requirements']):
        row['minimum_requirements'] = 'Not_Specified'

    if pd.isna(row['recommended_requirements']):
        row['recommended_requirements'] = 'Not_Specified'

    return row

df_cleaned = df.apply(handling_missing_data, axis=1)

df_cleaned.isna()
```

```python
columns_to_drop = ['achievements', 'original_price', 'discount_price', 'types']
df_cleaned = df_cleaned.drop(columns=[col for col in columns_to_drop if col in df_cleaned.columns])
```
