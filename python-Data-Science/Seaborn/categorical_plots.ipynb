```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
dataset = sns.load_dataset('titanic')

# Display dataset info
print("="*40)
print("üìå DATASET INFORMATION")
print("="*40)
print(dataset.info())

# Display first 5 rows
print("\n" + "="*40)
print("üîç FIRST 5 ROWS OF THE DATASET")
print("="*40)
print(dataset.head())

# Display last 5 rows
print("\n" + "="*40)
print("üìå LAST 5 ROWS OF THE DATASET")
print("="*40)
print(dataset.tail())

# Display basic statistics
print("\n" + "="*40)
print("üìä BASIC STATISTICS")
print("="*40)
print(dataset.describe())
```

```python
# Check for missing values
missing_values = dataset.isnull().sum()
print("\n" + "="*40)
print("‚ùó MISSING VALUES IN DATASET")
print("="*40)
print(missing_values)

# Dropping missing values
dataset_cleaned = dataset.dropna()

# Verify if missing values are handled
print("\n" + "="*40)
print("‚úÖ DATA CLEANED! (Missing values handled)")
print("="*40)
print(dataset_cleaned.isnull().sum())  
```

```python
#count plot
sns.countplot(data=dataset_cleaned,x='class',palette='coolwarm', legend=False, hue='class')
plt.title('Passangers Count by Class',fontsize=14)
plt.xlabel('Class',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.set_style('whitegrid')

sns.boxplot(data=dataset_cleaned, x='pclass', y='age', hue='pclass', palette='Set2', legend=False)

plt.title('Age Distribution by Passenger Class', fontsize=14)
plt.xlabel('Passenger Class', fontsize=12)
plt.ylabel('Age', fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.set_style('whitegrid')
sns.violinplot(data=dataset_cleaned, x='pclass', y='age', hue='sex', split=True, palette='husl')
plt.title('Age Distribution by Passenger Class', fontsize=14)
plt.xlabel('Passenger Class', fontsize=12)
plt.ylabel('Age', fontsize=12)
plt.legend(title="Sex")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.set_style('whitegrid')
sns.swarmplot(data=dataset_cleaned, x='pclass', y='age', hue='sex', palette='Set1', dodge=True)
plt.title('Swarm Plot of Age by Passenger Class', fontsize=14)
plt.xlabel('Passenger Class', fontsize=12)
plt.ylabel('Age', fontsize=12)
plt.legend(title="Sex")
plt.show()
```

```python
# Strip Plot
plt.figure(figsize=(8,6))
sns.set_style('whitegrid')
sns.stripplot(data=dataset_cleaned, x='pclass', y='age', jitter=True, palette='viridis', alpha=0.7,hue='pclass',legend=False)
plt.title('Age Distribution by Passenger Class (Strip Plot)', fontsize=14)
plt.xlabel('Passenger Class', fontsize=12)
plt.ylabel('Age', fontsize=12)
plt.show()
```

```python
# Point Plot
plt.figure(figsize=(8,6))
sns.set_style('whitegrid')
sns.pointplot(data=dataset_cleaned, x='pclass', y='age', hue='sex', palette='coolwarm', markers=['o', 's'], capsize=0.2)
plt.title('Average Age by Passenger Class and Gender', fontsize=14)
plt.xlabel('Passenger Class', fontsize=12)
plt.ylabel('Average Age', fontsize=12)
plt.legend(title='Gender')
plt.show()
```
