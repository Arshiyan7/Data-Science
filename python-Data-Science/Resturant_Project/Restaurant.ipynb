```python
#import libraries
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import data
df = pd.read_csv("resturant_data.csv")
df
```

```python
#check first five rows and last five rows and creating copy of dataFrame 
df_copy = df.copy()
print("first 5 rows: ",df.head(5))
print("\nlast 5 rows: ",df.tail(5))
```

```python
#removing / delimter from rating column 
df_copy['Rate'] = df_copy['Rate'].apply(lambda x:float(str(x).split('/')[0]))
df_copy.head()
```

```python
#Basic info of data frame
print(df.info())
print("\n")
print(df.describe())
```

```python
#Check for missing values
df.isna().sum()
df.fillna(0)
```

```python
#Count of Restaurants by Type
sns.countplot(x=df_copy['listed_in(type)'], 
              palette={'Buffet': 'red', 'Cake': 'green', 'Dine-out': 'blue', 'Takeaway': 'orange', 'Quick Bites': 'green', 'Cafe': 'maroon'}
              ,legend=False,hue=df_copy['listed_in(type)'])
plt.xlabel("Type of Restaurant")
```

```python
#Popularity of Restaurant Types Based on Votes
grouped_data = df_copy.groupby('listed_in(type)')['Votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result,marker='.',c='r')
plt.xlabel('Type of Resturant',c='black',size=12)
plt.ylabel('Votes',c='black',size=12)
```

```python
#Finding resturant with maximum votes
max_votes = df_copy['Votes'].max()
resturant_w_MaxVotes = df_copy.loc[df_copy['Votes']==max_votes,'Name']
print("Resturant with max votes is:\n",resturant_w_MaxVotes)
```

```python
#Ratings plot
sns.countplot(x=df_copy['Online Order'], hue=df_copy['Online Order'], 
              palette={'Yes': 'r', 'No': 'b'}, legend=False)
```

```python
#Distribution of Restaurant Ratings
plt.figure(figsize=(8, 5))
plt.hist(df_copy['Rate'], bins=8, color='teal', edgecolor='black', alpha=0.7)

plt.xlabel("Ratings", fontsize=12, color='darkred')
plt.ylabel("Frequency", fontsize=12, color='darkred')
plt.title("Ratings Distribution", fontsize=14, fontweight='bold', color='darkred')

plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

```python
#Cost Distribution for Dining Out
plt.figure(figsize=(10, 5))  
ax = sns.countplot(x=df_copy['approx_cost(for two people)'], color='darkslateblue')

plt.xticks(rotation=0.2, ha='right', fontsize=10)  
plt.xlabel("Approximate Cost (For Two People)", fontsize=12, color='darkblue')
plt.ylabel("Count", fontsize=12, color='darkblue')
plt.title("Distribution of Approximate Costs", fontsize=14, fontweight='bold', color='darkred')

plt.show()
```

```python
#Impact of Online Orders on Restaurant Ratings
plt.figure(figsize=(6,6))
sns.boxplot(data=df_copy,x='Online Order',y='Rate',palette={'Yes':'r','No':'Darkred'},hue=df_copy['Online Order'],legend=False)
```

```python
#Heatmap of Online Orders vs Restaurant Type
pivot_table = df_copy.pivot_table(index='listed_in(type)', columns='Online Order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Heatmap')
plt.xlabel('Online Order')
plt.ylabel('Listed In (Type)')
plt.show()
```

```python
#Analyzing data based on day and week  
#Convert date to datetime and extract days and week 
df_copy['Date'] = pd.to_datetime(df_copy['Date'], format='%d-%m-%Y')
df_copy['Day'] = df_copy['Date'].dt.day
df_copy['Week'] = df_copy['Date'].dt.isocalendar().week
```

```python
#visualizing data based on day and week  
#Day
day_group = df_copy.groupby('Day')['Votes'].sum()
day_group.plot(kind='line', marker='o', color='r', figsize=(8,4))
plt.show()
#Week
week_group = df_copy.groupby('Week')['Votes'].sum()
week_group.plot(kind='line', marker='*', color='b', figsize=(8,4))
plt.show()
```
