```python
import pandas as pd

data = {
    'Product': ['A', 'A', 'A', 'B', 'B', 'C', 'C', 'C'],
    'Region': ['North', 'South', 'East', 'North', 'South', 'North', 'South', 'East'],
    'Year': [2020, 2020, 2021, 2020, 2021, 2021, 2021, 2021],
    'Sales': [100, 150, 200, 120, 250, 300, 350, 400]
}

df = pd.DataFrame(data)
df
```

```python
piv_df = df.pivot(index=['Product','Year'],columns='Region',values="Sales")
piv_df
```

```python
#melt
import pandas as pd

data = {
    'Product': ['Laptop', 'Phone', 'Tablet'],
    'January': [150, 200, 120],
    'February': [180, 250, 150],
    'March': [210, 270, 170]
}

df = pd.DataFrame(data)
df
```

```python
df.melt(id_vars='Product',value_vars=['January','February','March'],var_name='')
```
